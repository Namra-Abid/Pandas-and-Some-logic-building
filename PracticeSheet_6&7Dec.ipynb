## Pivot Table


- Calculate the average turnover per category and year.  
- Calculate the sum of quantity per region and month.  
- Calculate the maximum cost per product grouped by region.  
- Calculate the total turnover per month for each category.  
- Calculate the minimum cost per year grouped by category and product.  

```python
# # Create a pivot table with filtering and aggregation
# pivot_df = pd.pivot_table(
#     countries_df,
#     index='continent',
#     columns='country',
#     values='gdp',
#     aggfunc='sum'
# )

# pivot_df
```

```python
sales.groupby(["Category","Year"])["Turnover"].mean()
```

```python
sales.pivot_table(index="Year",columns="Category",values="Turnover",aggfunc="mean")
```

### OR

```python
sales.pivot_table(index="Category",columns="Year",values="Turnover",aggfunc="mean")
```

- Calculate the sum of quantity per region and month.  


```python
sales.pivot_table(index="Region",columns="Month",values="Quantity",aggfunc="sum",)#.fillna(0)
```

- Calculate the maximum cost per product grouped by region.  


```python
sales.pivot_table(index="Region",columns="Product",values="Cost",aggfunc="max")
```

- Calculate the total turnover per month for each category.  


```python
sales.pivot_table(index="Category",columns="Month",values="Turnover",aggfunc="sum")
```

- Calculate the minimum cost per year grouped by category and product. 

```python
sales.pivot_table(index="Category",columns="Product",values="Cost",aggfunc="min")
```
